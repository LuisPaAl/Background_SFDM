```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
ym13 = np.loadtxt('data_m_e13')
ym11 = np.loadtxt('data_m_e11')
ym9 = np.loadtxt('data_m_e9')
ym7 = np.loadtxt('data_m_e7')
```

```python
a_i = 1.0e-9
lna = np.linspace(math.log(a_i),0., len(ym13[1,:]))

om_phi_m13 = lna*0.
om_phi_m11 = lna*0.
om_phi_m9 = lna*0.
om_phi_m7 = lna*0.

for i in range(len(lna)): 
    om_phi_m13[i] = -math.cos(ym13[1,i])
    om_phi_m11[i] = -math.cos(ym11[1,i])
    om_phi_m9[i] = -math.cos(ym9[1,i])
    om_phi_m7[i] = -math.cos(ym7[1,i])
```

```python
plt.plot(lna,ym13[2,:],"r--",label="$m^{-19} eV$")
plt.plot(lna,ym13[3,:],"r--")
plt.plot(lna,ym13[4,:],"r--")
plt.plot(lna,ym13[5,:],"r--")
plt.plot(lna,ym11[2,:],"k:",label="$m^{-21}eV$")
plt.plot(lna,ym11[3,:],"k:")
plt.plot(lna,ym11[4,:],"k:")
plt.plot(lna,ym11[5,:],"k:")
plt.plot(lna,ym9[2,:],"g",label="$m^{-23}eV$")
plt.plot(lna,ym9[3,:],"g")
plt.plot(lna,ym9[4,:],"g")
plt.plot(lna,ym9[5,:],"g")
plt.plot(lna,ym7[2,:],"b--",label="$m^{-25}eV$")
plt.plot(lna,ym7[3,:],"b--")
plt.plot(lna,ym7[4,:],"b--")
plt.plot(lna,ym7[5,:],"b--")
ann = plt.annotate('Baryons',xy=(-5,0.18),xytext=(-5,0.3),arrowprops=dict(arrowstyle='->'))
ann2 = plt.annotate('Dark energy',xy=(-1,0.35),xytext=(-6,0.6),arrowprops=dict(arrowstyle='->'))
ann3 = plt.annotate('Radiation',xy=(-12.3,0.8),xytext=(-18,0.75),arrowprops=dict(arrowstyle='->'))
ann4 = plt.annotate('Radiation',xy=(-12.2,0.95),xytext=(-18,0.75),arrowprops=dict(arrowstyle='->'))
ann5 = plt.annotate('SFDM',xy=(-12.3,0.2),xytext=(-18,0.1),arrowprops=dict(arrowstyle='->'))
ann6 = plt.annotate('SFDM',xy=(-11.5,0.13),xytext=(-18,0.1),arrowprops=dict(arrowstyle='->'))
plt.xlabel("$\ln (a)$")
plt.ylabel("")
plt.legend()
plt.show()
```

```python
plt.subplot(4,1,1)
plt.plot(lna,om_phi_m13,"r",label="$m^{-19} eV$")
plt.subplot(4,1,2)
plt.plot(lna,om_phi_m11,"k",label="$m^{-21} eV$")
plt.ylabel("$\omega_\phi$")
plt.legend()
plt.subplot(4,1,3)
plt.plot(lna,om_phi_m9,"g",label="$m^{-23} eV$")
plt.subplot(4,1,4)
plt.plot(lna,om_phi_m7,"b",label="$m^{-25} eV$")
plt.xlabel('$\ln (a)$')
plt.legend()
plt.show()

```
